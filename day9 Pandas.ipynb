The Grade Filter (Boolean Masking & Missing Data)

```python
import pandas as pd
```

```python
grades = pd.Series([85, None, 92, 45, None, 78, 55])

grades
```

```python
grades.isnull()
```

```python
filled_grades = grades.fillna(0)
filled_grades
```

```python
filtered_grades = filled_grades[filled_grades > 60]
filtered_grades
```

Task 1: The Integrity Audit (Missing Values & Duplicates)

```python
import pandas as pd
```

```python
data = pd.read_csv(r"C:\Users\archana\OneDrive\Desktop\customer_orders.csv")
print("Shape before cleaning:", data.shape)
```

```python
print("\nMissing Values Report:")
print(data.isna().sum())
```

```python
numeric_cols = data.select_dtypes(include=['number']).columns
for col in numeric_cols:
    data[col] = data[col].fillna(data[col].median())
```

```python
duplicates_before = data.duplicated().sum()
print("Duplicate rows found:", duplicates_before)
```

```python
data = data.drop_duplicates()
data
```

```python
print("Final Shape:", data.shape)
```

Task 2: The Type Fixer (Data Type Conversion)

```python
import pandas as pd
```

```python
data = pd.read_csv(r"C:\Users\archana\OneDrive\Desktop\customer_orders.csv")
print("Before Conversion:\n")
print(data.dtypes)
```

```python
data["amount"] = (
    data["amount"]
    .astype(str)
    .str.replace("$", "", regex=False)
    .str.replace(",", "", regex=False)
)

data["amount"] = pd.to_numeric(data["amount"], errors="coerce")

print("After Fixing Amount:\n")
print(data["amount"].dtype)

```

```python
data["order_id"] = pd.to_numeric(data["order_id"], errors="coerce")

print(data["order_id"].dtype)
```

```python
print("Final Data Types:\n")
print(data.dtypes)
```

Task 3: The Categorical Standardizer (String Cleaning)

```python
import pandas as pd

data = pd.DataFrame({
    "Location": [
        " New York",
        "new york",
        "NEW YORK ",
        "Los Angeles",
        " los angeles ",
        "LOS ANGELES"
    ]
})

data
```

```python
print("Unique values before cleaning:")
print(data["Location"].unique())
```

```python
data["Location"] = data["Location"].str.strip().str.title()
```

```python
print("Unique values after cleaning:")
print(data["Location"].unique())
```

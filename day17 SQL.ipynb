```python
import sqlite3

# Create or connect to database
conn = sqlite3.connect("internship.db")
cursor = conn.cursor()
```

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS interns (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT NOT NULL,
    track TEXT NOT NULL,
    stipend INTEGER
)
""")

conn.commit()
```

```python
cursor.executemany("""
INSERT INTO interns (name, track, stipend)
VALUES (?, ?, ?)
""", [
    ("Archana", "Data Science", 15000),
    ("Rahul", "Web Dev", 12000),
    ("Sneha", "AI/ML", 18000),
    ("Kiran", "Data Science", 16000),
    ("Meena", "Web Dev", 13000)
])

conn.commit()
```

```python
cursor.execute("SELECT name, track FROM interns")
result = cursor.fetchall()

for row in result:
    print(row)
```

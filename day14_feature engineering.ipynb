Task 1 :The Categorical Converter (Encoding)

```python
import pandas as pd
data = {
    'Transmission': ['Automatic', 'Manual', 'Automatic', 'Manual', 'Automatic'],
    'Color': ['Red', 'Blue', 'Green', 'Red', 'Blue']
}
df = pd.DataFrame(data)
print(df)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Transmission'] = le.fit_transform(df['Transmission'])
print(df)
```

```python
df = pd.get_dummies(df, columns=['Color'], drop_first=True)
print(df)
```

Task 2: The Leveling Field (Feature Scaling)

```python
import pandas as pd
df = pd.DataFrame({
    'Age': [22, 25, 30, 35, 40],
    'Salary': [20000, 30000, 50000, 80000, 120000]
})
print(df)
```

```python
from sklearn.preprocessing import StandardScaler
scaler_std = StandardScaler()
df_standardized = scaler_std.fit_transform(df)
df_standardized = pd.DataFrame(df_standardized, columns=df.columns)
print(df_standardized)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler_minmax = MinMaxScaler()
df_normalized = scaler_minmax.fit_transform(df)
df_normalized = pd.DataFrame(df_normalized, columns=df.columns)
print(df_normalized)
```

Task 3: The Complexity Creator (Polynomial Features)

```python
import numpy as np
import pandas as pd
np.random.seed(0)
X = np.linspace(-5, 5, 100).reshape(-1, 1)
y = X**2 + np.random.normal(0, 2, 100).reshape(-1, 1)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

model_linear = LinearRegression()
model_linear.fit(X, y)

y_pred_linear = model_linear.predict(X)
r2_linear = r2_score(y, y_pred_linear)

print("R² (Original Feature):", r2_linear)
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly.fit_transform(X)
```

```python
model_poly = LinearRegression()
model_poly.fit(X_poly, y)

y_pred_poly = model_poly.predict(X_poly)
r2_poly = r2_score(y, y_pred_poly)

print("R² (Polynomial Features):", r2_poly)
```
